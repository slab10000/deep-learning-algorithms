```python
import numpy as np
```

# Basic array operations

```python
my_list = [1,2,3,4,5]
print(f"my list is type {type(my_list)}")
my_array = np.array(my_list)
print(f"my array is type {type(my_array)}")
```

```python
my_float_array = np.array(my_list, "float32")
print(f"my float array is type {type(my_array)}")
print (my_float_array)
```

```python
my_list_of_lists = [my_list, my_list, my_list]
my_array_2d = np.array(my_list_of_lists)
print(my_array_2d)
```

```python
my_linspace_array = np.linspace(0,1,20)
print(my_linspace_array)
```

## Basic Matrix operations

```python
my_matrix = np.ones((5,6), dtype="int")
my_matrix
```

```python
my_list = [1,2,3,4,5]
my_list_of_lists = [my_list, my_list, my_list]
my_array_2d = np.array(my_list_of_lists)
print(my_array_2d)
```

```python
my_number_matrix = np.full((6, 7), 11)
print(my_number_matrix)
```

```python
sliced_matrix = my_number_matrix[:4,:2]
print(sliced_matrix)
```
